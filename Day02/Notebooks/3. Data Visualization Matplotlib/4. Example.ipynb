```python
import matplotlib.pyplot as plt
```

```python
import pandas as pd
```

```python
df = pd.read_csv('example.csv', skiprows = 1, usecols = [1,2,3], names = ['gdp_cap', 'life_exp', 'pop'])
```

```python
df.head()
```

```python
gdp_cap = df['gdp_cap']
life_exp = df['life_exp']
pop = df['pop']
```

```python
type(gdp_cap)
```

Setting up variables for code minimisation

xlabel, ylabel and title

```python
xlab = 'GDP per Capita [in USD]'
ylab = 'Life Expectancy [in years]'
title = 'World Development in 2007'
```

Let's plot it for once

```python
plt.plot(gdp_cap, life_exp)
plt.title(title)
plt.xlabel(xlab)
plt.ylabel(ylab)
```

Let's try scatter

```python
plt.scatter(gdp_cap, life_exp)
plt.title(title)
plt.xlabel(xlab)
plt.ylabel(ylab)
```

tick values for xticks

```python
tick_val = [1000,10000,100000]
tick_lab = ['1k','10k','100k']
```

```python
# Previous customisations
plt.scatter(gdp_cap, life_exp)
plt.title(title)
plt.xlabel(xlab)
plt.ylabel(ylab)

# Changing scale of x axis
plt.xscale('log')
plt.xticks(tick_val, tick_lab)
```

## Customising it further

```python
import numpy as np
```

```python
np_pop = np.array(pop)
```

```python
np_pop = 2 * np_pop
```

```python
#Changing size of bubbles
plt.scatter(gdp_cap, life_exp, s = np_pop)

# Previous customisations
plt.title(title)
plt.xlabel(xlab)
plt.ylabel(ylab)

plt.xscale('log')
plt.xticks(tick_val, tick_lab)
```
